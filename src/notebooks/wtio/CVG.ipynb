```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def analyze_complete_automotive_value_chain(file_path, min_intensity=0.001):
    """
    An√°lisis COMPLETO de todas las cadenas de valor upstream del sector autom√≥viles.
    VERSI√ìN CORREGIDA: Excluye componentes contables, solo sectores productivos reales.
    
    Parameters:
    -----------
    file_path : str
        Ruta al archivo WIOD
    min_intensity : float
        Intensidad m√≠nima para incluir en resultados (0.001 = 0.1%)
    """
    
    print("üöó AN√ÅLISIS COMPLETO: CADENAS DE VALOR REALES - AUTOM√ìVILES")
    print("="*75)
    
    data = pd.read_excel(file_path, engine='pyxlsb')
    print(f"‚úÖ Datos WIOD cargados: {data.shape}")
    
    # COMPONENTES CONTABLES A EXCLUIR (NO son sectores productivos)
    excluded_codes = ['II_fob', 'TOT', 'TXSP', 'EXP_adj', 'PURR', 'PURNR', 'VA', 'IntTTM', 'GO']
    excluded_descriptions = [
        'Total intermediate', 'Value added', 'Output at basic', 
        'taxes less subsidies', 'Transport Margins', 'Cif/ fob',
        'Direct purchases abroad', 'Purchases on the domestic'
    ]
    
    # 1. IDENTIFICAR SECTORES PRODUCTIVOS REALES (excluyendo componentes contables)
    productive_sectors = {}
    
    for i, row_id in enumerate(data.iloc[:, 0]):
        if pd.notna(row_id) and isinstance(row_id, str) and len(row_id) < 15:
            # Excluir c√≥digos contables
            is_excluded_code = any(excluded in str(row_id) for excluded in excluded_codes)
            
            if not is_excluded_code and row_id not in productive_sectors:
                # Obtener descripci√≥n
                description = str(data.iloc[i, 1]) if len(data.columns) > 1 else row_id
                
                # Excluir descripciones contables
                is_excluded_desc = any(excluded in description for excluded in excluded_descriptions)
                
                if not is_excluded_desc:
                    productive_sectors[row_id] = description
    
    print(f"üìä Sectores productivos identificados: {len(productive_sectors)}")
    print(f"üö´ Excluidos: {len(excluded_codes)} tipos de componentes contables")
    
    # 2. COLUMNAS DE AUTOM√ìVILES (solo uso intermedio real)
    auto_columns = []
    for col in data.columns:
        if str(col).startswith('C29.'):
            # Verificar que no sea componente contable
            is_excluded = any(suffix in str(col) for suffix in excluded_codes)
            if not is_excluded:
                auto_columns.append(col)
    
    print(f"üéØ Columnas autom√≥viles (uso intermedio): {len(auto_columns)}")
    
    # 3. CALCULAR DEMANDA TOTAL DE AUTOM√ìVILES (solo uso intermedio)
    total_auto_demand = 0
    for col in auto_columns:
        col_data = pd.to_numeric(data[col], errors='coerce').sum()
        total_auto_demand += col_data if pd.notna(col_data) else 0
    
    print(f"üí∞ Demanda intermedia total autom√≥viles: ${total_auto_demand:,.0f} millones")
    
    # 4. ANALIZAR CADA SECTOR UPSTREAM
    upstream_results = {}
    
    print(f"\nüîç Analizando {len(productive_sectors)} sectores upstream...")
    
    for sector_code, sector_description in productive_sectors.items():
        # INCLUIR C29 (autoconsumo del sector - cadenas intrasectoriales)
        # C29 ‚Üí C29 captura: componentes, partes, subensambles entre empresas automotrices
            
        # Buscar todas las filas de este sector (por pa√≠ses)
        sector_rows = []
        for i, row_id in enumerate(data.iloc[:, 0]):
            if pd.notna(row_id) and str(row_id) == sector_code:
                sector_rows.append(i)
        
        if not sector_rows:
            continue
        
        # Calcular flujos hacia autom√≥viles
        total_flow = 0
        country_flows = {}
        
        for row_idx in sector_rows:
            # Pa√≠s proveedor (columna 2)
            country = str(data.iloc[row_idx, 2])
            
            # Sumar flujos hacia todas las columnas C29 (uso intermedio)
            row_flow = 0
            row_data = data.iloc[row_idx]
            
            for col in auto_columns:
                if col in row_data.index:
                    val = pd.to_numeric(row_data[col], errors='coerce')
                    if pd.notna(val) and val > 0:
                        row_flow += val
            
            total_flow += row_flow
            if row_flow > 0:
                country_flows[country] = country_flows.get(country, 0) + row_flow
        
        # Calcular intensidad de dependencia
        intensity = total_flow / total_auto_demand if total_auto_demand > 0 else 0
        
        # Solo incluir si supera umbral m√≠nimo
        if intensity >= min_intensity and total_flow > 0:
            # Top 5 pa√≠ses proveedores
            top_suppliers = sorted(country_flows.items(), key=lambda x: x[1], reverse=True)[:5]
            
            upstream_results[sector_code] = {
                'name': sector_description,
                'intensity': intensity,
                'total_flow': total_flow,
                'top_suppliers': top_suppliers,
                'num_countries': len(country_flows)
            }
    
    print(f"‚úÖ Sectores upstream significativos: {len(upstream_results)}")
    
    return upstream_results

def create_clean_summary(results, top_n=30):
    """Crear resumen limpio con clasificaci√≥n por importancia econ√≥mica"""
    
    print(f"\n{'='*95}")
    print(f"üìã CADENAS DE VALOR COMPLETAS: TOP {top_n} SECTORES UPSTREAM DE AUTOM√ìVILES")
    print(f"{'='*95}")
    print("üîÑ INCLUYE: Autoconsumo sectorial (C29‚ÜíC29) - Cadenas intrasectoriales")
    print()
    
    # Ordenar por intensidad de dependencia
    sorted_results = sorted(results.items(), key=lambda x: x[1]['intensity'], reverse=True)
    
    # Clasificaci√≥n por nivel de criticidad
    categories = {
        'üî¥ CR√çTICOS (>2.0%)': [],
        'üü† MUY IMPORTANTES (1.0-2.0%)': [],
        'üü° IMPORTANTES (0.5-1.0%)': [],
        'üü¢ SIGNIFICATIVOS (0.1-0.5%)': [],
        'üîµ MENORES (0.05-0.1%)': [],
        '‚ö™ MARGINALES (<0.05%)': []
    }
    
    for sector_code, data in sorted_results[:top_n]:
        intensity = data['intensity']
        
        if intensity > 0.02:
            categories['üî¥ CR√çTICOS (>2.0%)'].append((sector_code, data))
        elif intensity > 0.01:
            categories['üü† MUY IMPORTANTES (1.0-2.0%)'].append((sector_code, data))
        elif intensity > 0.005:
            categories['üü° IMPORTANTES (0.5-1.0%)'].append((sector_code, data))
        elif intensity > 0.001:
            categories['üü¢ SIGNIFICATIVOS (0.1-0.5%)'].append((sector_code, data))
        elif intensity > 0.0005:
            categories['üîµ MENORES (0.05-0.1%)'].append((sector_code, data))
        else:
            categories['‚ö™ MARGINALES (<0.05%)'].append((sector_code, data))
    
    # Mostrar resultados por categor√≠as
    for category, sectors in categories.items():
        if sectors:
            print(f"\n{category}")
            print("-" * 80)
            
            for rank, (sector_code, data) in enumerate(sectors, 1):
                intensity_pct = data['intensity'] * 100
                flow_billions = data['total_flow'] / 1000
                
                # Acortar nombre del sector
                name = data['name']
                if len(name) > 55:
                    name = name[:52] + "..."
                
                # Destacar si es autoconsumo sectorial
                sector_note = ""
                if sector_code == 'C29':
                    sector_note = " üîÑ [AUTOCONSUMO SECTORIAL]"
                    name = "Motor vehicles - Internal supply chain (components, parts, subassemblies)"
                
                print(f"  {rank:2d}. {sector_code:<12} ‚îÇ {intensity_pct:5.2f}% ‚îÇ ${flow_billions:7.2f}B ‚îÇ {name}{sector_note}")
                
                # Mostrar top 3 pa√≠ses proveedores
                if data['top_suppliers']:
                    suppliers_str = " ‚îÇ ".join([f"{country}: ${flow/1000:.1f}B ({flow/data['total_flow']:.1%})" 
                                              for country, flow in data['top_suppliers'][:3]])
                    print(f"      üåç {suppliers_str}")
                    
    # Resumen estad√≠stico
    total_upstream_flow = sum(data['total_flow'] for data in results.values())
    avg_intensity = np.mean([data['intensity'] for data in results.values()])
    
    print(f"\nüìä RESUMEN ESTAD√çSTICO:")
    print(f"   ‚Ä¢ Total flujo upstream: ${total_upstream_flow/1000:.1f} mil millones")
    print(f"   ‚Ä¢ Intensidad promedio: {avg_intensity:.3%}")
    print(f"   ‚Ä¢ Sectores analizados: {len(results)}")

def plot_automotive_upstream_analysis(results, top_n=20):
    """Crear visualizaciones del an√°lisis upstream"""
    
    # Preparar datos para gr√°ficos
    sorted_results = sorted(results.items(), key=lambda x: x[1]['intensity'], reverse=True)
    top_results = dict(sorted_results[:top_n])
    
    sector_codes = []
    sector_names = []
    intensities = []
    flows = []
    
    for sector_code, data in sorted_results[:top_n]:
        sector_codes.append(sector_code)
        # Nombre corto para visualizaci√≥n
        name = data['name']
        if len(name) > 35:
            name = name[:32] + "..."
        sector_names.append(name)
        intensities.append(data['intensity'] * 100)
        flows.append(data['total_flow'] / 1000)  # En miles de millones
    
    # Crear figura con 4 subgr√°ficos
    fig = plt.figure(figsize=(22, 16))
    
    # 1. Intensidades de dependencia (gr√°fico principal)
    ax1 = plt.subplot(2, 2, 1)
    colors = plt.cm.tab20(np.linspace(0, 1, len(sector_codes)))
    bars = ax1.barh(range(len(sector_codes)), intensities, color=colors, alpha=0.8, edgecolor='black', linewidth=0.5)
    
    ax1.set_yticks(range(len(sector_codes)))
    ax1.set_yticklabels([f"{code}" for code in sector_codes], fontsize=10, fontweight='bold')
    ax1.set_xlabel('Intensidad de Dependencia (%)', fontsize=12, fontweight='bold')
    ax1.set_title('üöó Intensidad de Dependencia Upstream\n(Top 20 Sectores)', fontsize=14, fontweight='bold')
    ax1.grid(axis='x', alpha=0.3)
    ax1.invert_yaxis()
    
    # A√±adir valores en las barras
    for i, (bar, intensity) in enumerate(zip(bars, intensities)):
        ax1.text(bar.get_width() + 0.02, bar.get_y() + bar.get_height()/2, 
                f'{intensity:.2f}%', va='center', fontsize=9, fontweight='bold')
    
    # 2. Flujos monetarios
    ax2 = plt.subplot(2, 2, 2)
    bars2 = ax2.barh(range(len(sector_codes)), flows, color=colors, alpha=0.8, edgecolor='black', linewidth=0.5)
    
    ax2.set_yticks(range(len(sector_codes)))
    ax2.set_yticklabels([f"{code}" for code in sector_codes], fontsize=10, fontweight='bold')
    ax2.set_xlabel('Flujo Total (Miles de Millones USD)', fontsize=12, fontweight='bold')
    ax2.set_title('üí∞ Flujos Monetarios Upstream\n(Top 20 Sectores)', fontsize=14, fontweight='bold')
    ax2.grid(axis='x', alpha=0.3)
    ax2.invert_yaxis()
    
    # 3. Distribuci√≥n de intensidades (histograma)
    ax3 = plt.subplot(2, 2, 3)
    all_intensities = [data['intensity'] * 100 for data in results.values()]
    bins = np.logspace(np.log10(min(all_intensities)), np.log10(max(all_intensities)), 15)
    ax3.hist(all_intensities, bins=bins, alpha=0.7, color='lightblue', edgecolor='navy')
    ax3.set_xscale('log')
    ax3.set_xlabel('Intensidad de Dependencia (%) - Escala Log', fontsize=12, fontweight='bold')
    ax3.set_ylabel('N√∫mero de Sectores', fontsize=12, fontweight='bold')
    ax3.set_title('üìä Distribuci√≥n de Intensidades\n(Todos los Sectores)', fontsize=14, fontweight='bold')
    ax3.grid(alpha=0.3)
    
    # 4. Top pa√≠ses proveedores agregados
    ax4 = plt.subplot(2, 2, 4)
    
    # Agregar flujos por pa√≠s
    country_totals = {}
    for data in results.values():
        for country, flow in data['top_suppliers']:
            country_totals[country] = country_totals.get(country, 0) + flow
    
    top_countries = sorted(country_totals.items(), key=lambda x: x[1], reverse=True)[:12]
    countries, country_flows = zip(*top_countries)
    
    bars4 = ax4.bar(countries, [f/1000 for f in country_flows], 
                   alpha=0.8, color='coral', edgecolor='darkred', linewidth=1)
    ax4.set_ylabel('Flujo Total Agregado (Miles de Millones USD)', fontsize=12, fontweight='bold')
    ax4.set_title('üåç Top 12 Pa√≠ses Proveedores\n(Flujo Agregado)', fontsize=14, fontweight='bold')
    ax4.tick_params(axis='x', rotation=45, labelsize=11)
    ax4.grid(axis='y', alpha=0.3)
    
    # A√±adir valores en las barras del gr√°fico de pa√≠ses
    for bar in bars4:
        height = bar.get_height()
        ax4.text(bar.get_x() + bar.get_width()/2., height + height*0.01,
                f'{height:.1f}B', ha='center', va='bottom', fontweight='bold')
    
    plt.tight_layout(pad=3.0)
    plt.show()
    
    return fig

def generate_strategic_insights(results):
    """Generar insights estrat√©gicos sobre las cadenas de valor reales"""
    
    print(f"\n{'='*85}")
    print("üéØ INSIGHTS ESTRAT√âGICOS: VULNERABILIDADES Y OPORTUNIDADES")
    print("="*85)
    
    # Insight especial sobre autoconsumo C29
    if 'C29' in results:
        c29_data = results['C29']
        c29_intensity = c29_data['intensity']
        c29_flow = c29_data['total_flow']
        
        print(f"\nüîÑ AN√ÅLISIS ESPECIAL: AUTOCONSUMO SECTORIAL (C29 ‚Üí C29)")
        print("-" * 60)
        print(f"üìä Intensidad: {c29_intensity:.2%} | Flujo: ${c29_flow/1000:.1f}B")
        print(f"üîó Interpretaci√≥n: Cadenas intrasectoriales (componentes, partes, subensambles)")
        
        if c29_data['top_suppliers']:
            print(f"üåç Top pa√≠ses en cadenas intrasectoriales:")
            for i, (country, flow) in enumerate(c29_data['top_suppliers'][:5], 1):
                share = flow / c29_flow
                print(f"   {i}. {country}: ${flow/1000:.1f}B ({share:.1%})")
        print()
    
    # Continuar con an√°lisis normal...
    
    # 1. An√°lisis de concentraci√≥n geogr√°fica por pa√≠s
    country_analysis = {}
    for sector_code, data in results.items():
        for country, flow in data['top_suppliers']:
            if country not in country_analysis:
                country_analysis[country] = {
                    'sectors': set(),
                    'total_flow': 0,
                    'max_sector_share': 0,
                    'dominant_sectors': []
                }
            
            country_analysis[country]['sectors'].add(sector_code)
            country_analysis[country]['total_flow'] += flow
            
            # Calcular participaci√≥n en este sector
            sector_share = flow / data['total_flow']
            if sector_share > country_analysis[country]['max_sector_share']:
                country_analysis[country]['max_sector_share'] = sector_share
            
            if sector_share > 0.2:  # Dominancia >20%
                country_analysis[country]['dominant_sectors'].append((sector_code, sector_share))
    
    # Top pa√≠ses por diversificaci√≥n
    top_countries = sorted(country_analysis.items(), 
                          key=lambda x: (len(x[1]['sectors']), x[1]['total_flow']), 
                          reverse=True)[:8]
    
    print("\nüåç AN√ÅLISIS DE PROVEEDORES CLAVE:")
    print("-" * 60)
    for country, data in top_countries:
        sectors_count = len(data['sectors'])
        total_flow_b = data['total_flow'] / 1000
        max_share = data['max_sector_share']
        
        print(f"üá∫üá≥ {country}: {sectors_count} sectores | ${total_flow_b:.1f}B total | Max. dominio: {max_share:.1%}")
        
        if data['dominant_sectors']:
            dominant_str = ", ".join([f"{sector}({share:.1%})" for sector, share in data['dominant_sectors'][:3]])
            print(f"   ‚ö° Dominante en: {dominant_str}")
    
    # 2. Sectores con alta concentraci√≥n geogr√°fica (riesgo sist√©mico)
    risky_sectors = []
    for sector_code, data in results.items():
        if data['top_suppliers']:
            # Concentraci√≥n del top proveedor
            top_supplier_share = data['top_suppliers'][0][1] / data['total_flow']
            # Concentraci√≥n top 3
            top3_share = sum(flow for _, flow in data['top_suppliers'][:3]) / data['total_flow']
            
            if top_supplier_share > 0.4 or top3_share > 0.8:  # Alta concentraci√≥n
                risky_sectors.append((sector_code, data, top_supplier_share, top3_share))
    
    risky_sectors.sort(key=lambda x: x[2], reverse=True)  # Ordenar por concentraci√≥n
    
    print(f"\n‚ö†Ô∏è  SECTORES CON ALTA CONCENTRACI√ìN GEOGR√ÅFICA (RIESGO SIST√âMICO):")
    print("-" * 70)
    for sector_code, data, top1_share, top3_share in risky_sectors[:8]:
        intensity = data['intensity']
        top_country = data['top_suppliers'][0][0]
        
        print(f"üö® {sector_code}: {intensity:.2%} intensidad | {top_country} domina {top1_share:.1%}")
        print(f"   üìä Top 3 pa√≠ses: {top3_share:.1%} del mercado")
    
    # 3. Identificar sectores "sorprendentes" (no obvios) - excluyendo autoconsumo
    expected_critical = {'C24', 'C20', 'C22', 'C26', 'B', 'C25', 'C28', 'C29'}  # Incluir C29 como "esperado"
    
    sorted_results = sorted(results.items(), key=lambda x: x[1]['intensity'], reverse=True)
    surprising_sectors = []
    
    for sector_code, data in sorted_results:
        if sector_code not in expected_critical and data['intensity'] > 0.001:  # >0.1%
            surprising_sectors.append((sector_code, data))
    
    print(f"\nüîç SECTORES SORPRENDENTEMENTE IMPORTANTES:")
    print("-" * 50)
    for sector_code, data in surprising_sectors[:6]:
        intensity = data['intensity']
        flow_b = data['total_flow'] / 1000
        top_supplier = data['top_suppliers'][0][0] if data['top_suppliers'] else 'N/A'
        
        name = data['name']
        if len(name) > 50:
            name = name[:47] + "..."
            
        print(f"üí° {sector_code}: {intensity:.3%} | ${flow_b:.1f}B | Top: {top_supplier}")
        print(f"   üìù {name}")
    
    # 4. Patr√≥n de dominancia china
    china_dominance = []
    for sector_code, data in results.items():
        china_flow = 0
        china_share = 0
        
        for country, flow in data['top_suppliers']:
            if country == 'CHN':
                china_flow = flow
                china_share = flow / data['total_flow']
                break
        
        if china_share > 0.2:  # China >20% del sector
            china_dominance.append((sector_code, china_share, china_flow, data['intensity']))
    
    china_dominance.sort(key=lambda x: x[1], reverse=True)
    
    print(f"\nüá®üá≥ SECTORES BAJO DOMINANCIA CHINA (>20%):")
    print("-" * 55)
    china_total = sum(flow for _, _, flow, _ in china_dominance)
    
    for sector_code, share, flow, intensity in china_dominance[:10]:
        print(f"üî¥ {sector_code}: {share:.1%} mercado | ${flow/1000:.1f}B | {intensity:.2%} intensidad")
    
    print(f"\nüìà TOTAL FLUJO CHINA: ${china_total/1000:.1f} mil millones")
    print(f"üìä SECTORES DOMINADOS: {len(china_dominance)} de {len(results)}")

# FUNCI√ìN PRINCIPAL CORREGIDA
def run_complete_automotive_analysis(file_path, min_intensity=0.0005, top_display=25):
    """
    Ejecutar an√°lisis COMPLETO y LIMPIO de cadenas de valor automotrices.
    VERSI√ìN CORREGIDA: Excluye componentes contables.
    
    Parameters:
    -----------
    file_path : str
        Ruta al archivo WIOD
    min_intensity : float
        Intensidad m√≠nima para incluir sectores (0.0005 = 0.05%)
    top_display : int
        N√∫mero de sectores principales a mostrar
    """
    
    print("üöó AN√ÅLISIS COMPLETO Y LIMPIO: CADENAS DE VALOR AUTOMOTRICES")
    print("=" * 70)
    print("‚úÖ CORREGIDO: Solo sectores productivos reales (sin componentes contables)")
    print()
    
    # 1. An√°lisis completo limpio
    results = analyze_complete_automotive_value_chain(file_path, min_intensity)
    
    # 2. Resumen clasificado por importancia
    create_clean_summary(results, top_display)
    
    # 3. Visualizaciones mejoradas
    plot_automotive_upstream_analysis(results, min(20, len(results)))
    
    # 4. Insights estrat√©gicos
    generate_strategic_insights(results)
    
    print(f"\n‚úÖ AN√ÅLISIS COMPLETO TERMINADO")
    print(f"üìä Sectores upstream analizados: {len(results)}")
    print(f"üéØ Cobertura: CADENAS DE VALOR COMPLETAS (incluye autoconsumo C29‚ÜíC29)")
    print(f"üìà Umbral m√≠nimo aplicado: {min_intensity:.3%}")
    
    return results

# EJEMPLO DE USO
if __name__ == "__main__":
    file_path = '../../../src/data/external/WIOT2014_Nov16_ROW.xlsb'
    
    # Ejecutar an√°lisis completo limpio
    clean_results = run_complete_automotive_analysis(
        file_path, 
        min_intensity=0.0005,  # 0.05% m√≠nimo
        top_display=30         # Top 30 sectores
    )
```

```python
import pandas as pd
import numpy as np

def validate_wiod_structure(file_path):
    """
    Validar si estamos interpretando correctamente la estructura WIOD
    """
    
    print("üîç VALIDACI√ìN CR√çTICA: ¬øEstamos interpretando bien WIOD?")
    print("="*70)
    
    data = pd.read_excel(file_path, engine='pyxlsb')
    
    # TEST 1: ¬øHay patrones en las filas?
    print("\nüß™ TEST 1: Examinar patr√≥n de filas C24 (metales)")
    c24_rows = []
    for i, row_id in enumerate(data.iloc[:, 0]):
        if pd.notna(row_id) and str(row_id) == 'C24':
            c24_rows.append(i)
    
    print(f"Filas C24 encontradas: {c24_rows}")
    print(f"Diferencias entre filas: {[c24_rows[i+1]-c24_rows[i] for i in range(len(c24_rows)-1)]}")
    
    # TEST 2: ¬øQu√© hay en las columnas 1-3?
    print("\nüß™ TEST 2: ¬øQu√© contienen las columnas 1-3? (metadatos)")
    for i in c24_rows[:5]:
        row_data = data.iloc[i, :4]  # Primeras 4 columnas
        print(f"Fila {i}: {list(row_data)}")
    
    # TEST 3: ¬øLas columnas C29.X corresponden a pa√≠ses?
    print("\nüß™ TEST 3: Examinar headers de columnas C29")
    c29_cols = [col for col in data.columns if col.startswith('C29.')][:10]
    print("Primeras 10 columnas C29:")
    for i, col in enumerate(c29_cols):
        print(f"  {i+1:2d}. {col}")
    
    # TEST 4: ¬øLos valores hacen sentido econ√≥mico?
    print("\nüß™ TEST 4: Comparar valores entre pa√≠ses")
    print("Valores de C24 (metales) hacia C29.1, C29.7, C29.15 (AUS, CHN, DEU?):")
    
    for row_idx in c24_rows[:3]:
        values = []
        for col in ['C29.1', 'C29.7', 'C29.15']:
            if col in data.columns:
                val = pd.to_numeric(data.iloc[row_idx][col], errors='coerce')
                values.append(f"{col}=${val:,.0f}M" if pd.notna(val) else f"{col}=NaN")
        print(f"  Fila {row_idx}: {' | '.join(values)}")
    
    # TEST 5: ¬øHay informaci√≥n de pa√≠s en otras columnas?
    print("\nüß™ TEST 5: ¬øInformaci√≥n de pa√≠s en columnas 1-3?")
    for i in range(4, 20):  # Filas de datos reales
        row_info = [str(data.iloc[i, j]) for j in range(4)]
        print(f"Fila {i:2d}: {row_info}")
    
    # TEST 6: Verificar si la matriz es sim√©trica (pa√≠ses iguales en filas y columnas)
    print(f"\nüß™ TEST 6: Verificaci√≥n de simetr√≠a")
    print(f"Total filas: {data.shape[0]}")
    print(f"Total columnas: {data.shape[1]}")
    
    # Contar sectores √∫nicos
    unique_sectors = set()
    for row_id in data.iloc[:, 0]:
        if pd.notna(row_id) and isinstance(row_id, str) and len(row_id) < 10:
            unique_sectors.add(row_id)
    
    print(f"Sectores √∫nicos encontrados: {len(unique_sectors)}")
    print(f"Ejemplos: {sorted(list(unique_sectors))[:10]}")
    
    # TEST 7: ¬øCu√°ntos pa√≠ses por sector?
    sectors_to_check = ['C24', 'C29', 'A01']
    for sector in sectors_to_check:
        count = sum(1 for row_id in data.iloc[:, 0] if pd.notna(row_id) and str(row_id) == sector)
        print(f"Sector {sector}: {count} repeticiones (¬øpa√≠ses?)")
    
    return data

def check_country_mapping_hypothesis(file_path):
    """
    Verificar si nuestro mapeo de pa√≠ses es correcto
    """
    print("\n" + "="*70)
    print("üåç VERIFICACI√ìN: ¬øEl mapeo de pa√≠ses es correcto?")
    print("="*70)
    
    data = pd.read_excel(file_path, engine='pyxlsb')
    
    # Buscar filas C29 (autom√≥viles) y ver si los valores coinciden con lo esperado
    c29_rows = []
    for i, row_id in enumerate(data.iloc[:, 0]):
        if pd.notna(row_id) and str(row_id) == 'C29':
            c29_rows.append(i)
    
    print(f"Encontradas {len(c29_rows)} filas C29")
    
    # Examinar valores C29‚ÜíC29 (deber√≠a ser producci√≥n dom√©stica)
    print("\nüìä C29‚ÜíC29 (producci√≥n dom√©stica de autom√≥viles por pa√≠s):")
    expected_auto_producers = {
        'DEU': 'Alemania (alto)', 'JPN': 'Jap√≥n (alto)', 'USA': 'EE.UU. (alto)',
        'CHN': 'China (alto)', 'KOR': 'Corea (medio)', 'ITA': 'Italia (medio)',
        'FRA': 'Francia (medio)', 'GBR': 'Reino Unido (medio)'
    }
    
    for i, row_idx in enumerate(c29_rows):
        # Obtener valor C29‚ÜíC29 para este "pa√≠s"
        c29_domestic = data.iloc[row_idx]['C29'] if 'C29' in data.columns else 0
        c29_domestic = pd.to_numeric(c29_domestic, errors='coerce')
        
        if pd.notna(c29_domestic) and c29_domestic > 1000:  # Solo valores significativos
            print(f"  Fila {row_idx} (pa√≠s #{i+1}): ${c29_domestic:,.0f}M producci√≥n dom√©stica")
            
            # Si esto representa Alemania/Jap√≥n/EE.UU., deber√≠a ser muy alto
            if c29_domestic > 50000:
                print(f"    ‚≠ê ALTO - Probablemente pa√≠s productor importante")
            elif c29_domestic > 10000:
                print(f"    üî∏ MEDIO - Productor secundario")
            else:
                print(f"    ‚ö™ BAJO - No es productor principal")

# EJECUTAR VALIDACI√ìN COMPLETA
if __name__ == "__main__":
    file_path = '../../../src/data/external/WIOT2014_Nov16_ROW.xlsb'
    
    # Validar estructura
    data = validate_wiod_structure(file_path)
    
    # Verificar mapeo de pa√≠ses
    check_country_mapping_hypothesis(file_path)
    
    print("\n" + "="*70)
    print("ü§î CONCLUSI√ìN: Basado en estos tests, ¬ølos resultados tienen sentido?")
    print("Si Croacia aparece como mayor productor, hay un ERROR en nuestro c√≥digo.")
    print("="*70)
```
